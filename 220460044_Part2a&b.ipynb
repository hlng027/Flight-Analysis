## Part 2 (a)

```python
# Delay by time of day
for i in range(1991, 2001):
    ontime = pd.read_csv(f"{i}.csv")
    
    # Convert 'DepTime' column to string type
    ontime['DepTime'] = ontime['DepTime'].astype(str)

    # Replace non-numeric values with NaN
    ontime['DepTime'] = pd.to_numeric(ontime['DepTime'].str[:2], errors='coerce')

    # Calculate TimeOfDayGroup
    ontime['TimeOfDayGroup'] = pd.cut(ontime['DepTime'],
                                  bins=[-float('inf'), 6, 12, 18, float('inf')],
                                  labels=["Night", "Morning", "Afternoon", "Evening"])
    
    # Group by TimeOfDayGroup and calculate average delay for each year
    daydelay = ontime.groupby('TimeOfDayGroup')['ArrDelay'].mean().reset_index().sort_values(by='ArrDelay')

    # Sort daydelay in ascending order based on 'ArrDelay'
    daydelay_sorted = daydelay.sort_values(by='ArrDelay', ascending=True)

    # Print the sorted DataFrame
    print(daydelay_sorted)
```

```python
# Delay by days of week
for i in range(1991, 2001):
    ontime = pd.read_csv(f"{i}.csv")
    
    # Convert 'DepTime' column to string type
    ontime['DepTime'] = ontime['DepTime'].astype(str)

    # Replace non-numeric values with NaN
    ontime['DepTime'] = pd.to_numeric(ontime['DepTime'].str[:2], errors='coerce')
    
    # Group by DayOfWeek and calculate average delay
    weekdelay = ontime.groupby('DayOfWeek')['ArrDelay'].mean().reset_index().sort_values(by='ArrDelay')
    
    # Sort weekdelay in ascending order based on 'ArrDelay'
    weekdelay_sorted = weekdelay.sort_values(by='ArrDelay', ascending=True)
    
    print(weekdelay_sorted)
```

```python
# Delay by month
for i in range(1991, 2001):
    ontime = pd.read_csv(f"{i}.csv")
    
    # Convert 'DepTime' column to string type
    ontime['DepTime'] = ontime['DepTime'].astype(str)

    # Replace non-numeric values with NaN
    ontime['DepTime'] = pd.to_numeric(ontime['DepTime'].str[:2], errors='coerce')
    
    # Group by Month and calculate average delay
    monthdelay = ontime.groupby('Month')['ArrDelay'].mean().reset_index().sort_values(by='ArrDelay')
    
    # Sort monthdelay in ascending order based on 'ArrDelay'
    monthdelay_sorted = weekdelay.sort_values(by='ArrDelay', ascending=True)
    
    print(monthdelay_sorted)
```

```python
# Read csv file
planes_df = pd.read_csv("plane-data.csv")
```

```python
results = {} #Initialize empty dictionary to store results

for i in range(1991, 2001):
    planes_year = planes_df.copy()
    planes_year['ManufacturingDate'] = pd.to_datetime(planes_year['issue_date'], format='%m/%d/%Y', errors='coerce')
    planes_year['PlaneAge'] = i - planes_year['ManufacturingDate'].dt.year
    planes_year = planes_year[planes_year['PlaneAge'] >= 0]
    ontime_age = pd.merge(ontime, planes_year[['tailnum', 'PlaneAge']], left_on='TailNum', right_on='tailnum')

    # Drop rows with missing values in the 'DepDelay' and 'ArrDelay' columns
    ontime = ontime.dropna(subset=['DepDelay', 'ArrDelay'])
    
    agedelay_year = ontime_age.groupby('PlaneAge')[['DepDelay', 'ArrDelay']].mean().reset_index()
    agedelay_year['AvgTotalDelay'] = agedelay_year['DepDelay'] + agedelay_year['ArrDelay']

    results[str(i)] = agedelay_year
```

```python
# Print the results
for i in range(1991, 2001):
    print(results[str(i)])
```
